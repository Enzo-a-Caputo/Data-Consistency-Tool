```python
import sys, os
sys.path.insert(0, os.path.join(os.getcwd(), ".."))

import pandas as pd
import numpy as np
from consistency import run_app
```

# Load Level Data

```python
file_path = r'C:\Projetos\Consistencia\SalvosCajuru\Nivel\UHE Cajuru Montante\Nivel.xlsx'

df_level = pd.read_excel(file_path, skiprows=0, usecols=[0, 1], names=['Date', 'Level'])
df_level['Date'] = pd.to_datetime(df_level['Date'])
df_level = df_level.drop_duplicates(subset='Date').reset_index(drop=True)

df_level.head()
```

# Load Rainfall Data

```python
file_path = r'C:\Projetos\Consistencia\ChuvaContador\Contador_Comutado2044105.xlsx'

df = pd.read_excel(file_path, sheet_name='SérieHorária', skiprows=1)
df_rain = df.iloc[:, [1, 6]].copy()
df_rain.columns = ['Date', 'Counter']
df_rain['Date'] = pd.to_datetime(df_rain['Date'])

# Compute rainfall increment (difference between consecutive counter readings)
df_rain['Increment'] = df_rain['Counter'].diff()

df_rain.head()
```

# Run Consistency Tool

```python
run_app(
    df_level,
    df_rain,
    col_primary='Level',
    col_secondary='Increment',
    date_col='Date',
    port=8054,
)
```
